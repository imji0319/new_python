# MNIST 손글씨 숫자 

## 숫자 데이터의 구조 시각화

```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml('MNIST_784')
mnist.data.shape
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

```python
fig, ax = plt.subplots(6,8, subplot_kw = dict(xticks =[], yticks =[]))

for i, axi in enumerate(ax.flat):
    axi.imshow(mnist.data[1250*i].reshape(28,28), cmap = 'gray_r') # 각 이미지는 28*28 구성되어 있음
```

```python
# 데이타 1/30 만 사용. ( 전체 데이터를 계산하려면 시간 소요 큼 )

data = mnist.data[::30]
target = mnist.target[::30]
```

```python
from sklearn.manifold import Isomap
```

```python
model = Isomap(n_components = 2)
proj = model.fit_transform(data)

plt.figure(figsize = (13,10))
plt.scatter(proj[:,0], proj[:,1], c=target.astype(np.int64), cmap = plt.cm.get_cmap('jet', 10))
plt.colorbar(ticks = range(10))
plt.ylim(-10000,10000)
plt.xlim(-10000,10000)
plt.clim(-0.5, 9.5)
```

```python
# 사영할 숫자 '1' 의 1/4 을 선택 

data = mnist.data[mnist.target == '1'][::4]
```

```python
data
```

```python
data.shape
```

```python
from matplotlib import offsetbox
def plot_components(data, model, images = None, ax = None, thumb_frac = 0.05, cmap = 'gray') :
    ax = ax or plt.gca()
    
    proj = model.fit_transform(data)
    ax.plot(proj[:,0], proj[:,1], '.k')
    
    if images is not None :
        min_dist_2 = (thumb_frac * max(proj.max(0) -proj.min(0))) ** 2
        shown_images = np.array([2 * proj.max(0)])
        
        for i in range(data.shape[0]):
            dist = np.sum((proj[i] - shown_images) ** 2, 1)
            
            if np.min(dist) < min_dist_2 :
                
                # 너무 가까운 점들은 보여주지 말것 
                continue
                
            shown_images = np.vstack([shown_images, proj[i]])
            imagebox = offsetbox.AnnotationBbox(offsetbox.OffsetImage(images[i], cmap = cmap), proj[i])
            
            ax.add_artist(imagebox)
```

```python
fig, ax = plt.subplots(figsize = (10,10))

model = Isomap(n_neighbors = 5, n_components = 2, eigen_solver ='dense')
plot_components(data, model, images= data.reshape((-1, 28,28)),
                ax = ax, thumb_frac = 0.05, cmap ='gray_r')
```
