# 안면 인식 파이프라인

## 유향 경사 히스토그램 (HOG, Histogram of Oriented Gradients)
이미지 픽세을 빛 같은 혼동 요인에 관계 없이 광범위하게 정보를 제공하는 임지 특징에 민감한 벡터 표현으로 변환 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import seaborn as sns; sns.set()
```

## HOG 단계
1. 선택적으로 이미지를 사전에 정규화 한다. 이로써 빛의 변화에 의존하지 않는 특징을 얻게 된다.
2. 수평적, 수직적 밝기 경도에 민감한 두 필터로 이미지를 휘감는다. 이 필터가 모서리와 윤곽, 질감 정보를 포착한다.
3. 이미지를 사전 정의한 크기의 셀로 나누고 각 셀 내의 경사 방향과 히스토그램을 계산한다.
4. 이웃하는 셀의 블록과 비교해서 각 셀의 히스토그램을 계산한다.
5. 각 셀의 정보로부터 1차원 특징 벡터를 구성한다. 

```python
from skimage import data, color, feature
import skimage.data
```

```python
image = color.rgb2gray(data.chelsea())
hog_vec , hog_vus = feature.hog(image, visualize=True)
```

```python
fig, ax = plt.subplots(1, 2, figsize = (12, 6),
                      subplot_kw = dict (xticks =[], yticks =[]))

ax[0].imshow(image, cmap = 'gray')
ax[0].set_title('input image')

ax[1].imshow(hog_vus)
ax[1].set_title('visualization of HOG features')
```
