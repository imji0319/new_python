### 나이브 베이즈 분류 Naive Bayes classification 
매우 높은 차원의 데이터세트에 적합한 상당히 빠르고 간단한 분류 알고리즘.  
조정 가능한 모수가 매우 적기 때문에 간편한 기준선으로 쓰기에 매우 유용.   

#### 베이즈 분류 
통계량의 조건부 확률 사이의 관계를 나타내는 방정식인 베이즈 정리 기반.   
관측된 특징 feature가 주어졌을 때 레이블 label의 확률을 구하는 데 관심이 있음.   
<i>P(L|features)</i>로 쓸 수 있음 
![베이즈 정리](./bayes.png)

두 개의 레이블 사이에서 결정하려고 할 때, 각 레이블의 사후 확률의 비율을 계산.
![베이즈 정리2](./bayes2.png)

이제 각 레이블에 대한 <i>P(features/L)</i>를 계산할 수 있는 모델이 필요 : 생성 모델 Generative Model   
각 레이블에 대한 이 생성 모델 지정은 베이즈 분류기를 훈련시키는 주요 부분.  
단순화된 가정들을 사용함으로써 작업이 간단해졌고 그로 인해 'Naive' 라는 말을 사용.  


#### 가우스 나이브 베이즈 Gaussian Naive Bayes 
각 레이블의 데이터가 간단한 가우스 분포로부터 추출된다고 가정한는 것부터 시작.   
데이터가 차원 사이에 공분산이 없는 가우스 분포를 따른다고 가정함에 따라

## K-평균 알고리즘  

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 10, random_state =0)
clusters = kmeans.fit_predict(digits.data)

kmeans.cluster_centers_.shape
```

```python
fig, ax = plt.subplots(2,5, figsize = (8,3))
centers = kmeans.cluster_centers_.reshape(10,8,8)

for axi, center in zip(ax.flat, centers):
    axi.set(xticks = [], yticks =[])
    axi.imshow(center, interpolation = 'nearest', cmap = plt.cm.binary)
```

## 나이브하지 않은 베이즈 Not So Naive Bayes

```python
from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.neighbors import KernelDensity
from sklearn.model_selection import GridSearchCV
```

```python
class KDEClassifier(BaseEstimator, ClassifierMixin):
    
    """
    KDE 기반 베이즈 생성 분류 모수 ----
    - 대역폭 : 부동 소수점, 각 클래스 내 커널 대역폭 
    - 커널 : 문자열, kernelDensity 에 전달할 커널 이름 
    """
    
    def __init__(self, bandwidth = 1.0, kernel = 'gaussian'):
        self.bandwidth = bandwidth
        self.kernel = kernel
        
    def fit(self, X, y):
        self.classes_ = np.sort(np.unique(y))
        training_sets = [X[y == yi] for yi in self.classes_]
        self.models_ = [KernelDensity(bandwidth = self.bandwidth, kernel = self.kernel).fit(Xi) for Xi in training_sets]
        
        self.logpriors_ = [np.log(Xi.shape[0] / X.shape[0])
                          for Xi in training_sets]
        
        return self 
    def predict_proba(self, X):
        logprobs = np.array([model.score_samples(X) for model in self.models_]).T
        result = np.exp(logprobs + self.logpriors_)
        return result/result.sum(1, keepdims = True)
        
    def predict(self, X):
        return self.classes_[np.argmax(self.predict_proba(X), 1)]
            
```

```python
bandwidths = 10 ** np.linspace(0, 2, 100)
grid = GridSearchCV(KDEClassifier(), {'bandwidth' : bandwidths})
grid.fit(digits.data, digits.target)

scores = [val for val in grid.cv_results_["mean_test_score"]]
```

기존의 코드는    
scores = [val.mean_validation_score for val in grid.grid_score_] 

해당 코드는 여러 에러가 발생하였음.   
grid_score 는 sklearn의 업데이트 이후 cv_results_ 로 변경 되었으며, 해당 함수의의 Attribute 안에는 mean_validation_score 값이 없음 .  
해당 코드로 mean_test_score로만 변경한 것 역시 val를 str로 처리 되어 attribute를 가져올 수 없음.  
for 문 안에 cv_results_ 의 구조가 dictionary 구조임을 활용하여 해당 값을 가져옴. 

```python
plt.semilogx(bandwidths, scores)
plt.xlabel('bandwidth')
plt.ylabel('accuracy')
plt.title('KDE Model Performance')

print(grid.best_params_)
print('accuracy = ', grid.best_score_)
```
