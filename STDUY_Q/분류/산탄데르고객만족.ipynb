# 산탄데르 고객 만족 예측 

산탄데르 고객 만족 데이터를 통해 고객 만족 여부를 예측하라.   
평가지표는 f1-score를 활용할 것이며   
테스트 데이터에 대해 예측한 후 예측 데이터를 제출하라. 

```python
import pandas as pd
```

```python
train = pd.read_csv('../../DS_Series/DS_ML_guide/data/santander/train.csv')
test = pd.read_csv('../../DS_Series/DS_ML_guide/data/santander/test.csv')
```

```python
train['var3'].replace(-999999, 2, inplace =True)
```

```python
from sklearn.metrics import f1_score
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
train.drop(['ID'], axis = 1, inplace =True)
test.drop(['ID'], axis = 1, inplace = True)
```

```python
x = train.drop(['TARGET'], axis = 1)
y = train['TARGET']
```

```python
y.value_counts()
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.25, 
                                                   random_state = 619 )
```

```python
from lightgbm import LGBMClassifier
```

```python
lgbm = LGBMClassifier(random_state = 619,
                     n_estimators = 1000, learning_rate = 0.003)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
lgbm.fit(train_x, train_y,
        eval_set = [(train_x, train_y), (test_x, test_y)], eval_metric ='F1',
        early_stopping_rounds = 100, verbose = 100)
```

```python
pred = lgbm.predict(test_x)
```

```python
f1_score(test_y, pred)
```

```python
lgbm = LGBMClassifier(random_state = 619, learning_rate = 0.003 )
```

```python
params = {'num_leaves' : [32,64],
         'max_depth' :[128,160],
         'min_child_samples' : [60,100],
         'subsample' : [0.8, 1],
         'n_estimators' : [1000, 2000, 3000]}
```

```python
gridCV = GridSearchCV(lgbm, param_grid = params)
```

```python
gridCV.fit(train_x, train_y,
          early_stopping_rounds = 100,
          eval_set = [(train_x, train_y), (test_x, test_y)],
           eval_metric ='F1',verbose = 1000)
```
