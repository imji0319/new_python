# 피마 인디언 당뇨병 예측 

피마 인디언 당뇨병 데이터 세트를 활용하여 당뇨병 여부를 판단하는 머신 모델을 구축하라.   
테스트 데이터에 대해 평가지표 ROC_AUC를 적용하여 평가할 것이다. 

```python
import pandas as pd
```

```python
data = pd.read_csv('../../DS_Series/DS_ML_guide/data/diabetes.csv')
```

```python
data
```

```python
target = data['Outcome']
```

```python
dataset = data.drop(['Outcome'], axis = 1)
```

```python
dataset.describe()
```

```python
dataset.isna().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x = scaler.fit_transform(dataset)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, target, test_size = 0.25, 
                                                   random_state = 0)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
lr.fit(train_x, train_y)
```

```python
pred = lr.predict(test_x)
```

```python
roc_auc_score(test_y, pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(train_x, train_y)
```

```python
pred = rf.predict(test_x)
```

```python
roc_auc_score(test_y, pred)
```

```python
rf2 = RandomForestClassifier(n_estimators = 500,
                            random_state = 0)
```

```python
rf2.fit(train_x, train_y)
```

```python
pred = rf2.predict(test_x)
```

```python
roc_auc_score(test_y, pred)
```

```python
from xgboost import XGBClassifier
```

```python
help(XGBClassifier)
```

```python
xgbc = XGBClassifier(n_estimators = 300,
                    learning_rate = 0.08,
                    random_state = 0)
```

```python
xgbc.fit(train_x , train_y,eval_metric = 'logloss')
```

```python
pred = xgbc.predict(test_x)
```

```python
roc_auc_score(test_y, pred)
```
