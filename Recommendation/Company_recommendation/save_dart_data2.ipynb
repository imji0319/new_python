### 2020년 기업별 사업보고서
상장한 기업이 아니더라도 일정 수준의 수익이 발생할 경우 사업보고서를 제출할 의무가 있음.    
비상장법인이더라도 1. 모집 또는 매출실적이 있거나, 
2, 외부감사대상 법인으로 증권의 소유자가 500인 이상인 발행인은 사업보고서를 제출하여야 합니다.   
사업보고서는 각 사업연도 경과 후 90일 이내, 분,반기보고서는 그 기간 경과 후 45일 이내에 제출하여야 합니다.


=> 사업보고서가 있다는 것은 일정 수준 이상의 수익이 발생하고 있음을 의미하고 이는 일정부분의 기업의 안정성을 의미하기도 함.

```python
report_2020[report_2020.corp_code.isin(check_c)].sort_values('corp_code').report_nm.value_counts()
```

### 기업별 사업보고서 dict 생성 

```python
comp_report_dict = {}
for i,v in enumerate(report_2020.name_code):
    doc = report_2020.loc[i, 'rcept_no']
    try:
        text = dart.document(doc)
    except : 
        print(doc)
    soup = BeautifulSoup(text)
    doc_text = soup.get_text()
    
    # 정규식 처리 
    doc_text = re.sub('[\n-]', '', doc_text)
    doc_text = re.sub('&cr;', '', doc_text)
    
    comp_report_dict[v] = doc_text
```
