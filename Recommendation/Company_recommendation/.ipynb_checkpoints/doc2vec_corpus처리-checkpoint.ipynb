```python
from gensim.models.doc2vec import TaggedDocument
from gensim.models import Doc2Vec
from konlpy.tag import Mecab
# Konlpy 의 Mecab 형태소 분석기
mecab = Mecab()
```

```python
class Doc2VecCorpus:
    def __init__(self, fname):
        self.fname = fname

    def __iter__(self):
        with open(self.fname, encoding = 'utf-8') as f :
            for doc in f:
                company, text = doc.split('\t')
                yield TaggedDocument(words = mecab.morphs(text),
                                     tags = [company])
```

```python
import pandas as pd
```

```python
f = open("dataset/report_2020_business.txt", 'r')
line = f.readline()
```

```python
mecab.morphs(line.split('\t')[1])
```

```python
mecab.nouns(line.split('\t')[1])
```

```python
doc2vecCorpus = Doc2VecCorpus('dataset/report_2020_business.txt')
```

```python
model = Doc2Vec(epochs=20,
                    negative=5)
model.build_vocab(doc2vecCorpus)
```

```python
%%time
# train model
model.train(doc2vecCorpus,
            total_examples= model.corpus_count,
            epochs=model.epochs)
```

```python
model.save('business_report_doc2vec_model')
```
