## 사업보고서 데이터 추출 
정시 공시 목록 데이터를 확인한 결과 [첨부정정]사업보고서의 경우 해당 보고서 접수번호로는 document 조회가 불가능함. 마지막 보고서가 첨부정정이 아니도록 데이터 프레임을 구성하기 위해 `final= False` 로 설정하여 중간변경 보고서까지 수집해 확인작업이 필요함. 

```python
import OpenDartReader
import configparser
import pandas as pd 
import re
import warnings

warnings.filterwarnings('ignore')
config = configparser.ConfigParser()
config.read('config.ini', encoding ='utf-8')
```

```python
api_key = config['dart']['key']
dart = OpenDartReader(api_key)
```

```python
%%time
temp = pd.concat([dart.list(start= '2021-01-01' ,end = '2021-03-01', kind ='A',final=False),
                  dart.list(start= '2021-03-02' ,end = '2021-06-02', kind ='A',final=False),
                  dart.list(start= '2021-06-03' ,end = '2021-09-03', kind ='A',final=False)],ignore_index = True)
```

```python
%%time
temp = pd.concat([temp, 
                  dart.list(start= '2021-09-04' ,end = '2021-12-04', kind ='A',final=False),
                  dart.list(start= '2021-12-04' ,end = '2021-12-31', kind ='A',final=False),
                  dart.list(start= '2022-01-01', end = '2022-03-20', kind ='A',final=False)],ignore_index = True)
```

```python
temp
```

### 2021년 사업보고서 
2021년 사업보고서에는 2022년 3월 20일 기준으로 아직 제출하지 않은 기업이 있을 수 있음. 3월 말까지 기한이 있기 때문.    
또한 제출한 사업보고서에 대해 3월 이후에 계속해서 수정 및 정정이 가능해 완전한 보고서가 아닐 가능성이 잇음. 

```python
report_2021 = temp[temp['report_nm'].str.contains(r'사업보고서(.)*(2021.12)')]
```

### 2020년 사업보고서

```python
report_2020 = temp[temp['report_nm'].str.contains(r'사업보고서(.)*(2020.12)')]
```

```python
report_2020
```

```python
# 연장보고서 제거 
report_2020 = report_2020[~report_2020.report_nm.str.contains('연장')]
```

```python
# 첨부정정 제거 
report_2020 = report_2020[~report_2020.report_nm.str.contains('첨부정정')]
```

### 중간 변경 보고서 관련 EDA
기업별로 제출한 보고서의 개수를 계산한 결과 가장 많이 제출한 경우은 6번.

```python
report_2020.groupby('corp_code')['rcept_no'].count().to_frame().sort_values('rcept_no')
```
