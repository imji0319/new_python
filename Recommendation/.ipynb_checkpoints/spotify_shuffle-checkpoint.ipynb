```python
df = pd.merge(df, song_meta, how ='left', left_on ='song_id',right_on ='id').drop('id', axis = 1)
```

```python
genre_all = load_json(data_path+'genre_gn_all.json')
```

```python
df['genre_name'] = df['song_gn_gnr_basket'].apply(lambda x : [genre_all[i] for i in x])
```

```python
df = df[['song_id','song_name','artist_name_basket','genre_name','genre_dtl_name']]
```

```python
df
```

### Genre 기준으로 shuffle 

```python
from collections import Counter
```

```python
genre = Counter(sum(df['genre_name'].tolist(), []))
```

```python
df['main_genre'] = df['genre_name'].apply(lambda x : x[0])
```

#### Logical Block 형성

```python
genre_song = {}
for i in range(len(df)):
    temp = df.iloc[i,:]
    s_id = temp['song_id']
    gn = temp['main_genre']
    genre_song.setdefault(gn, [])
    genre_song[gn].append(s_id)
```

```python
genre_song
```

```python
{k : len(v) for k,v in genre_song.items()}
```

#### 빈공간을 포함한 리스트 생성

```python
for i in genre_song:
    if len(genre_song[i]) < 54:
        temp = ['-']*(54- len(genre_song[i]))
        genre_song[i].extend(temp)
```

#### 같은 블록 내 Shuffle 
이때 random shuffle 이 활용된다. 

```python
import random 
```

```python
list(map(random.shuffle, genre_song.values()))
```

```python
np.array(genre_song['일렉트로니카'])
```

#### 블록의 세로 영역을 기준으로 정렬

```python
np.array(list(genre_song.values()))[:,0]
```

```python
temp = np.array(list(genre_song.values()))
```

```python
shuffle_list = []

for i in range(len(temp[0,:])):
    t = list(filter(lambda x : x != '-',temp[:,i]))
    shuffle_list.extend(t)
```

```python
shuffle_list = list(map(int, shuffle_list))
```

```python
shuffle_list
```

```python
shuffle_df = df.set_index('song_id').loc[shuffle_list]
```
