### t-SNE

t-Stochastic Nearest Neighbor (t-SNE) 는 vector visualization을 위해서 자주 이용되는 알고리즘.  
t-SNE는 고차원 데이터를 2차원의 지도로 표현하며, 벡터 시각화를 위한 다른 알고리즘보다 안정적인 임베딩 학습 결과를 보여준다. 이는 t-SNE가 데이터 간 거리를 stochastic probability로 변환하여 임베딩에 이용하기 때문이다. 

```python
from sklearn.manifold import TSNE
```

```python
m = TSNE(learning_rate = 50)
```

```python
tsne_feature = m.fit_transform(movies_tr_num)
```

```python
df = pd.DataFrame()
```

```python
tsne_feature
```

```python
df['x'] = tsne_feature[:,0]
df['y'] = tsne_feature[:,1]
```

```python
df = pd.concat([movies_tr_num, df], axis = 1)
```

#### K-Fold 교차검증 
K-Fold = 5 

```python
from sklearn.model_selection import KFold
k_fold = KFold(n_splits = 5, shuffle = True)

for train_idx, val_idx in k_fold.split(X_train):
    print(len(train_idx), len(val_idx))
```

```python
lgbmr_model = lgb.LGBMRegressor()
```

```python
models = []

for train_idx, val_idx in k_fold.split(X_train):
    x_t = X_train.iloc[train_idx]
    y_t = y_train.iloc[train_idx]
    
    x_val = X_train.iloc[val_idx]
    y_val = y_train.iloc[val_idx]
    
    models.append(lgbmr_model.fit(x_t, y_t, eval_set = (x_val, y_val),
                                 early_stopping_rounds = 100, verbose = 100 ))
```

```python
models
```

```python
preds = [] 

for model in models:
    
    preds.append(model.predict(test_du))
```

```python
len(preds)
```

```python
kfoldLGBMR_du = submission.copy()
kfoldLGBMR_du['box_off_num'] = np.mean(preds, axis = 0 )
kfoldLGBMR_du.head()
```

```python
kfoldLGBMR_du.to_csv('kfoldLGBMR_du.csv', index= False)
```
