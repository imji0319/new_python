### 지역 라벨링 

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit(train.지역)
label = le.transform(train.지역)
```

```python
le.classes_
```

```python
T.지역 = le.transform(T.지역)
```

```python
T
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
y = T['등록차량수']
x = T.drop(['등록차량수'], axis = 1)
```

```python
sc_x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(sc_x, y,
                                                   test_size = 0.25,
                                                   random_state = 0)
```

```python
len(train_x)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(random_state = 0)
```

```python
rfr.fit(train_x, train_y)
```

```python
pred = rfr.predict(test_x)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(test_y, pred)
```

```python
from catboost import CatBoostRegressor
catb = CatBoostRegressor(
         loss_function='MAE',
         n_estimators=500, 
         learning_rate=0.05, 
         random_state=42,
    verbose = 100
    )
```

```python
catb.fit(train_x, train_y)
```

```python
pred = catb.predict(test_x)
```

```python
mean_squared_error(test_y, pred)
```

```python
SupplyType['공공분양'] = pd.Series([0 for i in range(len(SupplyType.index))], index = SupplyType.index)
SupplyType['장기전세'] = pd.Series([0 for i in range(len(SupplyType.index))], index = SupplyType.index)
SupplyType['공공임대(5년)'] = pd.Series([0 for i in range(len(SupplyType.index))], index = SupplyType.index)
```

```python
SupplyType = SupplyType[['공공분양','공공임대(10년)','공공임대(50년)','공공임대(5년)','공공임대(분납)','국민임대','영구임대','임대상가','장기전세','행복주택']]
```

```python
AO = test.groupby(['단지코드','자격유형'])['공가수'].count().unstack().fillna(0)
```

```python
AO['B'] = pd.Series([0 for i in range(len(AO.index))], index = AO.index)
AO['F'] = pd.Series([0 for i in range(len(AO.index))], index = AO.index)
AO['O'] = pd.Series([0 for i in range(len(AO.index))], index = AO.index)
```

```python
AO = AO[['A','B','C','D','E','F','G','H','I','J','K','L','M','N','O']]
```

```python
AreaSize = test.groupby(['단지코드',pd.cut(test.전용면적, bins)]).size().unstack()
AreaSize.columns = ['면적10','면적20','면적30','면적40','면적50','면적60',
                          '면적70','면적80','면적90']

AreaPP = test.groupby(['단지코드',pd.cut(test.전용면적, bins)])['전용면적별세대수'].mean().unstack().fillna(0)
AreaPP.columns = ['세대수10','세대수20','세대수30','세대수40','세대수50','세대수60','세대수70','세대수80','세대수90']
```

```python
test.replace('-',0, inplace =True)
```

```python
test[['임대보증금','임대료']] = test[['임대보증금','임대료']].fillna(0)
test[['임대보증금','임대료']] = test[['임대보증금','임대료']].astype(int)
Money = test.replace(0,np.NaN).groupby('단지코드')[['임대보증금','임대료']].mean()  # 가격을 알 수 없는 것들을 제외시켜서 계산
```

```python
Money.fillna(0,inplace = True)
```

```python
col = ['단지코드','총세대수','건물구분','지역','공급유형','전용면적','전용면적별세대수',
       '공가수','자격유형','임대보증금','임대료','지하철','버스','단지내주차면수']
```

```python
test.columns = col
```

```python
ori.지하철 = ori.지하철.fillna(0)
ori.버스 = ori.버스.fillna(0)
```

```python
ori_code = ori.drop_duplicates()
```

```python
ori_code.set_index('단지코드', inplace =True)
```

```python
# ori_code + AreaSize
T = pd.merge(ori_code, AreaSize, 
             left_index =True, 
            right_index = True) 

# ori_code + AreaSize + SupplyType
T = pd.merge(T, SupplyType,
        left_index =True,
        right_index = True)

# ori_code + AreaSize + SupplyType + AO
T = pd.merge(T, AO,
        left_index = True, 
        right_index =True)

# ori_code + AreaSize + SupplyType + AO + AreaPP
T = pd.merge(T, AreaPP,
        left_index =True,
        right_index = True)

# ori_code + AreaSize + SupplyType + AO + AreaPP + Money
T = pd.merge(T,  Money,
        left_index = True,
        right_index = True)
```

```python
T.지역 = le.transform(T.지역)  # train 데이터로 라벨링 학습한 모델 사용 
```
