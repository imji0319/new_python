## 임대보증금과 임대료는 단지코드별 평균처리 

```python
train.replace('-',0, inplace =True)
```

```python
train[train['임대료']=='-']
```

```python
train[['임대보증금','임대료']] = train[['임대보증금','임대료']].fillna(0)
```

```python
train[['임대보증금','임대료']].isna().sum()
```

```python
train[['임대보증금','임대료']] = train[['임대보증금','임대료']].astype(int)
```

```python
train.isna().sum()
```

```python
Money = train.replace(0,np.NaN).groupby('단지코드')[['임대보증금','임대료']].mean()  # 가격을 알 수 없는 것들을 제외시켜서 계산
```
