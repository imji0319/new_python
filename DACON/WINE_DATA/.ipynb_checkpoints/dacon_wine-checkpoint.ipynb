## Outlier 처리

```python
import numpy as np 

def remove_outlier_test(d_cp, column):
    
    for i in column:
        
        fraud_col_data = d_cp[i]
    
        Q1 = np.percentile(fraud_col_data.values, 25)
        Q3 = np.percentile(fraud_col_data.values, 75)
        
        IQR = Q3 - Q1
        outlier_step = 1.5* IQR
        lowest = Q1 - outlier_step
        highest = Q3 + outlier_step
        
        outlier_index = fraud_col_data[(fraud_col_data < lowest) | (fraud_col_data > highest)]
    
        #print(len(outlier_index), outlier_index.index, sep='\n')
        d_cp.drop(outlier_index.index, axis = 0, inplace =True )
    
    print(d_cp.shape)

    return d_cp
```

```python
# Type 변수 ( white, red ) - > 0, 1 로 변환 
train['type'] = train['type'].map({'white': 0, 'red' : 1}).astype(int)
test['type'] = test['type'].map({'white':0, 'red': 1}).astype(int)
```

```python
outlier_col = ['residual sugar','free sulfur dioxide','density']
train_out = remove_outlier_test(train, outlier_col)
train_out.head()
```
