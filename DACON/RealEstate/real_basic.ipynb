#### KOSPI지수와 비교 

```python
from pandas_datareader import data as pdr
import yfinance as yf

yf.pdr_override()
```

```python
kospi = pdr.get_data_yahoo('^KS11', '1976-01-01')
```

```python
kospi.head(2)
```

```python
kospi.reset_index(inplace = True)
```

```python
kospi['연월'] = pd.to_datetime(kospi['Date']).dt.strftime('%Y-%m')
```

```python
kospi.head(2)
```

```python
kospi['주식규모'] = kospi['Adj Close'] * kospi['Volume']
```

```python
kospi1997= kospi[kospi['Date'].dt.year > 1996][['Adj Close','주식규모','연월']]
```

```python
grouped = kospi1997.groupby(['연월'])
```

```python
kospiadjClose = grouped.mean()
```

```python
kospiadjClose = kospiadjClose.reset_index()
```

```python
kospiadjClose.head(5)
```

```python
kospiadjClose.shape
```

```python
# 연월별 총수주액 데이터 셋과 kospi 평균 데이터 셋 결합 

tmp = 국내건설수주액[국내건설수주액['연월'].dt.year > 1996][['연월','총수주액[백만원]']]
```

```python
tmp['연월'] = pd.to_datetime(tmp['연월']).dt.strftime('%Y-%m')
```

```python
tmp.head(3)
```

```python
temp = pd.merge(kospiadjClose, tmp, on = '연월', how ='inner')
```

```python
temp['연월'] = pd.to_datetime(temp['연월'], format = '%Y-%m')
```

```python
temp.set_index('연월', inplace = True)
```

```python
temp['총수주액[만원]'] = temp['총수주액[백만원]']* 100
temp['총수주액[십만원]'] = temp['총수주액[백만원]']* 10
temp['총수주액[천만원]'] = temp['총수주액[백만원]']/ 10
```

```python
temp.info()
```

```python
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates


fig, ax = plt.subplots(figsize = (25,7))
plt.plot('총수주액[만원]', data = temp, label ='국내건설 총수주액[만원]')
plt.plot('주식규모', data = temp, label = '대한민국 주식시장(Adj Close * Volumn)')
plt.legend(fontsize = 12)


date_form = DateFormatter("%Y-%m")
ax.xaxis.set_major_formatter(date_form)
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=30))
plt.xticks(rotation = 45)


plt.xlabel('연월')
plt.title('국내건설 총 수주액과 국내 주식시장의 관계 ', size = 20, pad = 12)
```
