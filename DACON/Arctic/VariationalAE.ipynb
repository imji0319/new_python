```python
ice_250 = ice_250.reshape(len(ice_250), 448, 304, 1)
```

```python
def autoencoder(img_shape, code_size) : 
    
    # encoder 
    
    encoder = Sequential()
    encoder.add(InputLayer(img_shape))
    
    encoder.add(Conv2D(filters = 32, kernel_size= 3, strides = 1, padding = "same"))
    encoder.add(LeakyReLU())
    encoder.add(Conv2D(filters = 64, kernel_size= 3, strides = 2, padding = "same"))
    encoder.add(LeakyReLU())
    encoder.add(Conv2D(filters = 64, kernel_size= 3, strides = 2, padding = "same"))
    encoder.add(LeakyReLU())
    encoder.add(Conv2D(filters = 64, kernel_size= 3, strides = 1, padding = "same"))
    encoder.add(LeakyReLU())
    
    encoder.add(Flatten())
    encoder.add(Dense(code_size[0]*code_size[1]))
    encoder.add(Reshape(code_size))
    
    
    # decoder 
    
    decoder = Sequential()
    decoder.add(InputLayer(code_size))
    decoder.add(Flatten())
    decoder.add(Dense(np.prod(img_shape)))
    decoder.add(Reshape(img_shape))
    
    decoder.add(Conv2DTranspose(64, 3, 1, padding ="same"))
    decoder.add(LeakyReLU())
    decoder.add(Conv2DTranspose(64, 3, 2, padding ="same"))
    decoder.add(LeakyReLU())
    decoder.add(Conv2DTranspose(32, 3, 2, padding ="same"))
    decoder.add(LeakyReLU())
    decoder.add(Conv2DTranspose(1, 3, 1, padding ="same"))
    decoder.add(Activation("sigmoid"))
    
    
    return encoder, decoder
```

```python
image_shape = ice_250.shape[1:]
code_shape = (28, 19)
```

```python
encoder, decoder = autoencoder(image_shape, code_shape)
```

```python
inp = Input(image_shape)
code = encoder(inp)
recon = decoder(code)

auto = Model(inp, recon)
```
