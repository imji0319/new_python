# Prophet parameters 조정 
페이스북에서 제공하는 Prophet 라이브러리를 사용하여 각 구분별 모델을 생성해 데이터 학습 및 예측을 진행한 모델에 대해 DACON 의 제출한 결과 : 0.1569 

성능을 보다 향상시키기 위해 `add_seasonality()` 함수를 활용해 보고자 함

```python
import numpy as np
import pandas as pd 
import datetime
import matplotlib.pyplot as plt 
%matplotlib inline 

plt.rc('font', family = 'AppleGothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('한국가스공사_시간별 공급량_20181231.csv', encoding = 'euc-kr')
```

```python
test = pd.read_csv('가스공급량 수요예측모델개발_data/test.csv')
sub = pd.read_csv('가스공급량 수요예측모델개발_data/sample_submission.csv')
```

```python
df['시간_수정'] = df['시간'] -1 
df['시간_수정'] = df['시간_수정'].astype(str)
```

```python
def tohour(val):
    if len(val) == 1:
        return '0'+val+':00:00'
    else:
        return val+':00:00'
```

```python
df['시간_수정'] = df['시간_수정'].apply(tohour)
```

```python
df['연월일시간'] = df['연월일']+' '+df['시간_수정']
df['연월일시간'] = pd.to_datetime(df['연월일시간'])
```

```python
A = df[df['구분'] =='A']
B = df[df['구분'] =='B']
C = df[df['구분'] =='C']
D = df[df['구분'] =='D']
E = df[df['구분'] =='E']
G = df[df['구분'] =='G']
H = df[df['구분'] =='H']
```

```python
A['공급량'].plot()
```

```python
A = A[['연월일시간','공급량']]
A.columns = ['ds','y']

B = B[['연월일시간','공급량']]
B.columns = ['ds','y']

C = C[['연월일시간','공급량']]
C.columns = ['ds','y']

D = D[['연월일시간','공급량']]
D.columns = ['ds','y']

E = E[['연월일시간','공급량']]
E.columns = ['ds','y']

G = G[['연월일시간','공급량']]
G.columns = ['ds','y']

H = H[['연월일시간','공급량']]
H.columns = ['ds','y']
```

```python
import prophet
```

```python
a_model = prophet.Prophet(changepoint_prior_scale = 0.3, 
                          yearly_seasonality = 20, 
                          weekly_seasonality = 15)
b_model = prophet.Prophet(changepoint_prior_scale = 0.3, 
                          yearly_seasonality = 20, 
                          weekly_seasonality = 15)
c_model = prophet.Prophet(changepoint_prior_scale = 0.3, 
                          yearly_seasonality = 20, 
                          weekly_seasonality = 15)
d_model = prophet.Prophet(changepoint_prior_scale = 0.3, 
                          yearly_seasonality = 20, 
                          weekly_seasonality = 15)
e_model = prophet.Prophet(changepoint_prior_scale = 0.3, 
                          yearly_seasonality = 20, 
                          weekly_seasonality = 15)
g_model = prophet.Prophet(changepoint_prior_scale = 0.3, 
                          yearly_seasonality = 20, 
                          weekly_seasonality = 15)
h_model = prophet.Prophet(changepoint_prior_scale = 0.3, 
                          yearly_seasonality = 20, 
                          weekly_seasonality = 15)
```

```python
a_model.fit(A)
b_model.fit(B)
c_model.fit(C)
d_model.fit(D)
e_model.fit(E)
g_model.fit(G)
h_model.fit(H)
```

```python
test['일자'] =test['일자|시간|구분'].str.slice(0,10)
test['시간'] = test['일자|시간|구분'].str.slice(11,13)
test['구분'] = test['일자|시간|구분'].str.slice(14)
test['시간_수정'] = test['시간'].astype(int)-1
test['시간_수정'] =test['시간_수정'].astype(str)
test['시간_수정'] = test['시간_수정'].apply(tohour)
test['연월일시간'] = test['일자'] +' '+ test['시간_수정']
test['연월일시간'] = pd.to_datetime(test['연월일시간'])
```

```python
a_test = test[test['구분'] =='A'][['연월일시간']]
a_test.columns =['ds']

b_test = test[test['구분'] =='B'][['연월일시간']]
b_test.columns =['ds']

c_test = test[test['구분'] =='C'][['연월일시간']]
c_test.columns =['ds']

d_test = test[test['구분'] =='D'][['연월일시간']]
d_test.columns =['ds']

e_test = test[test['구분'] =='E'][['연월일시간']]
e_test.columns =['ds']

g_test = test[test['구분'] =='G'][['연월일시간']]
g_test.columns =['ds']

h_test = test[test['구분'] =='H'][['연월일시간']]
h_test.columns =['ds']
```

```python
a_forecast = a_model.predict(a_test)
b_forecast = b_model.predict(b_test)
c_forecast = c_model.predict(c_test)
d_forecast = d_model.predict(d_test)
e_forecast = e_model.predict(e_test)
g_forecast = g_model.predict(g_test)
h_forecast = h_model.predict(h_test)
```

```python
model_list = [a_model, b_model, c_model, d_model, e_model, g_model, h_model]
fore_list = [a_forecast, b_forecast, c_forecast,d_forecast, e_forecast, g_forecast, h_forecast]
type_list = ['A','B','C','D','E','G','H']
```

```python
fig, axes = plt.subplots(7,1, figsize = (15,15))

for idx , ax in enumerate(axes.flat):
    model_list[idx].plot(fore_list[idx], ax = ax)
    ax.set_title(type_list[idx]+' 모델 Prophet 그래프')
```

```python
a_model.plot_components(a_forecast)
plt.show()
```

```python
tmp = test[test['구분']== 'A']
a_forecast.index = tmp.index
a_tmp = pd.concat([tmp, a_forecast['yhat']],axis = 1)

tmp = test[test['구분']== 'B']
b_forecast.index = tmp.index
b_tmp = pd.concat([tmp, b_forecast['yhat']],axis = 1)


tmp = test[test['구분']== 'C']
c_forecast.index = tmp.index
c_tmp = pd.concat([tmp, c_forecast['yhat']], axis = 1)


tmp = test[test['구분']== 'D']
d_forecast.index = tmp.index
d_tmp = pd.concat([tmp, d_forecast['yhat']],axis = 1)

tmp = test[test['구분']== 'E']
e_forecast.index = tmp.index
e_tmp = pd.concat([tmp, e_forecast['yhat']],axis = 1)


tmp = test[test['구분']== 'G']
g_forecast.index = tmp.index
g_tmp = pd.concat([tmp, g_forecast['yhat']],axis = 1)

tmp = test[test['구분']== 'H']
h_forecast.index = tmp.index
h_tmp = pd.concat([tmp, h_forecast['yhat']],axis = 1)

total_test = pd.concat([a_tmp, b_tmp, c_tmp, d_tmp, e_tmp, g_tmp, h_tmp])
```

```python
total_test['yhat']
```

```python
sub['공급량'] = total_test['yhat']
```

```python
sub.to_csv('submission/prophet_구분모델2.csv', index = False)
```
