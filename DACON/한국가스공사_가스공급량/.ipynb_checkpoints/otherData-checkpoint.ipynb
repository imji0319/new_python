```python
import numpy as np
import pandas as pd
```

```python
elect  = pd.read_csv('한국전력거래소_시간별전력수요량.csv', encoding ='euc-kr')
```

```python
elect
```

```python
df = pd.read_csv('한국가스공사_시간별 공급량_20181231.csv', encoding ='euc-kr')
```

```python
df[:3]
```

```python
elect_df = elect.set_index('날짜').stack().to_frame().reset_index()
```

```python
elect_df.columns = ['연월일', '시간', '전력수요량']
```

```python
elect_df.시간 = elect_df.시간.str.replace('시','')
```

```python
df.시간 = df.시간.astype(str)
```

```python
elect_df[:3]
```

```python
data = df
```

```python
data['연월일'] = pd.to_datetime(data['연월일'])
data['year'] = data['연월일'].dt.year
data['month'] = data['연월일'].dt.month
data['day'] = data['연월일'].dt.day
data['weekday'] = data['연월일'].dt.weekday
```

```python
d_map ={}
for i, d in enumerate(data['구분'].unique()):
    d_map[d] = i 
data['구분'] = data['구분'].map(d_map)
```

```python
test = pd.read_csv('가스공급량 수요예측모델개발_data/test.csv')
sub = pd.read_csv('가스공급량 수요예측모델개발_data/sample_submission.csv')
test['일자'] = test['일자|시간|구분'].str.split(' ').str[0]
test['시간'] = test['일자|시간|구분'].str.split(' ').str[1].astype(int)
test['구분'] = test['일자|시간|구분'].str.split(' ').str[2]
test['일자'] = pd.to_datetime(test['일자'])
test['year'] = test['일자'].dt.year
test['month'] = test['일자'].dt.month
test['day'] = test['일자'].dt.day
test['weekday'] = test['일자'].dt.weekday
test['구분'] = test['구분'].map(d_map)
```

```python
elect_df
```

```python
data['연월일'] = data['연월일'].astype(str)
```

```python
train_data = pd.merge(data,elect_df, how = 'left', right_index = False)
```

```python
train_data[:5]
```

```python
train_data.isna().sum()
```

```python
train_data.year = train_data.year.astype(int)
train_data.month = train_data.month.astype(int)
train_data.day = train_data.day.astype(int)
train_data.weekday = train_data.weekday.astype(int)
train_data.구분 = train_data.구분.astype(int)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
y = train_data['공급량'] 
x = train_data[['전력수요량','구분','year','month','day','weekday']]
```

```python
scaler = MinMaxScaler(feature_range = (0,1))
scaler.fit(pd.DataFrame(x['전력수요량']))
```

```python
x['전력수요량'] = scaler.transform(pd.DataFrame(x['전력수요량']))
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, shuffle = True, 
                                                    random_state = 1000)
```

```python
import lightgbm as lgb
```

```python
d_train = lgb.Dataset(X_train, y_train)
d_test = lgb.Dataset(X_test, y_test)
```

```python
params = {'objective' : 'regression',
         'metric':'mae',
         'seed': 1000}
```

```python
lgbr = lgb.train(params, d_train, 10000, d_test, verbose_eval = 1000,
                  early_stopping_rounds = 1000)
```

```python
test_x = test[['일자','시간','구분','year','month','day','weekday']]
test_x.columns = ['연월일','시간','구분','year','month','day','weekday']
```

```python
test_x['연월일'] = test_x['연월일'].astype(str)
```

```python
test_x.info()
```

```python
test_x['시간'] = test_x['시간'].astype(str)
```

```python
test_x = pd.merge(test_x, elect_df, how = 'left', right_index = False)
```

```python
test_x
```

```python
test_x = test_x[['전력수요량','구분','year','month','day','weekday']]
```

```python
test_x['전력수요량'] = scaler.transform(pd.DataFrame(test_x['전력수요량']))
```

```python
pred = lgbr.predict(test_x)
```

```python
sub['공급량'] = pred
```

```python
sub.to_csv('submission/lgbm_elecData2.csv', index = False)
```
